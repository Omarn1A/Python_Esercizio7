```python
total_Coderre=df['Coderre'].sum()
total_Bergeron=df['Bergeron'].sum()
total_Joly=df['Joly'].sum()
total = pd.DataFrame({
    'Candidate': ['Coderre', 'Bergeron', 'Joly'],
    'Total Votes': [total_Coderre, total_Bergeron, total_Joly]
})
total.plot(kind='bar', x='Candidate')
```
